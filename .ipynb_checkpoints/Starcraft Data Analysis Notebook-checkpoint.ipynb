### EDA

When I have a new dataset I always start by reading the data dictionary. I was pretty familiar with many starcraft terms but had to look up what a PAC (perception action cycle) was, which I found here: https://tl.net/forum/starcraft-2/401425-starcraft-2-science-skillcraft-results.

Afterwards I always just take an overall look at the dataframe to see if there is anything unusual and I found that the Age, HoursPerWeek, and TotalHours columns were objects when they should be intergers (or numerical columns). What I found is that missing data was input as '?' strings as opposed NaNs.

```python
starcraft.info() # take an overall view of the dataframe
```

```python
starcraft
```

```python
#check to see if there are any other weird values in the column, 
#I checked the other columns as well but for brevity sake, they aren't included in the notebook
starcraft["Age"].unique()
```

```python
# Make Player Ranks more human readable
rank_dict = {1:"Bronze", 2:"Silver", 3:"Gold", 4:"Plat", 5:"Diamond", 6:"Master", 7:"GrandMaster", 8:"Pro"}
starcraft["LeagueName"] = starcraft["LeagueIndex"].apply(lambda x: rank_dict[x])
```

```python
 # check to see if missingness is seemingly random or biased
starcraft[(starcraft["Age"] == "?") | (starcraft["HoursPerWeek"] == "?") | (starcraft["TotalHours"] == "?")][["LeagueName",
        "Age", "HoursPerWeek", "TotalHours"]]
```

Looking at the distributions of the missing data in the age and both hours columns gives me information to know whether I should impute the data with some value like the median, drop the rows, or handle it some other way. 

Because almost all of the missingness is in pro players with a couple diamond players slightly in the mix as well, it means I cannot really impute the data (especially for the pro players) with any value because it would not be a good estimate. Because the missing values are biased towards pro players, this means I cannot base my model heavily on any of these three variables if I want to be able to predict pro players with accuracy. 

I could build a model where if the age is missing it guesses 'pro rank' but I would assume that isn't helpful for our goal and then the model would be useless if we did have the age of a pro player. We only want to train models on information we do have, not on information we do not.


```python
# Next, I check the distribution of the ranks for discrepancies
starcraft["LeagueName"].value_counts()
```

```python
#somewhat bell-shaped curve, with more data would probably seem even more normal
starcraft["LeagueIndex"].plot(kind="hist") 
```

Next, I want to look at the correlation so I can explore the most important variables.


I have some knowledge of starcraft, and I expect that higher APM, Hotkey usage, and of course hours played should be some of the highest correlates to rank but let's see.

```python
starcraft.drop(columns="GameID").corr()[["LeagueIndex"]].sort_values(by="LeagueIndex")
```

```python
# the same thing but with pro players removed so we can see the correlation of age and hours columns
no_mi_data = starcraft[starcraft["TotalHours"] != "?"].drop(columns="GameID")
no_mi_data[["Age", "TotalHours", "HoursPerWeek"]] = no_mi_data[["Age", "TotalHours", "HoursPerWeek"]].astype('int64')
no_mi_data.corr()[["LeagueIndex"]].sort_values(by="LeagueIndex") 
```

I was wrong about total hours, as it isn't very heavily correlated at all with rank, as well as age and hours per week being relatively less impactful with slight correlations so I feel pretty good about dropping columns those from the model completely.

Let's look into the 6 strongest correlations (all absolute value of .4 corr or higher), which are: ActionLatency, GapBetweenPACs, APM, NumberOfPACs, AssignToHotkeys, and SelectByHotkeys

```python
avg_by_rank = starcraft.groupby("LeagueName").mean().sort_values(by="LeagueIndex", ascending=True)[["ActionLatency",
                                "GapBetweenPACs", "APM", "NumberOfPACs", "AssignToHotkeys", 
                                "SelectByHotkeys", "LeagueIndex"]]
avg_by_rank
```

We can see very clear lines here with ActionLatency and	GapBetweenPACs going down with rank and the other 4 going up with rank.

This makes sense because starcraft is a game with an important emphasis on micromanagement and multi-tasking. So the variables about latency and delay reduce with higher rank and the variables about quickness and number of actions in short periods of times increase with rank.

```python
# graphs for visual analysis
plt.plot(avg_by_rank.index, avg_by_rank["ActionLatency"], label="ActionLatency")
plt.plot(avg_by_rank.index, avg_by_rank["APM"], label="APM")
plt.plot(avg_by_rank.index, avg_by_rank["GapBetweenPACs"], label="GapBetweenPACs")
plt.legend()
plt.title("High Correlation Variable Differences by Rank")
plt.xlabel("Rank")
plt.ylabel("Units")
plt.show()
```

```python
# graphs for visual analysis, different scale for these variables
plt.plot(avg_by_rank.index, avg_by_rank["AssignToHotkeys"], label="AssignToHotkeys", color="Purple")
plt.plot(avg_by_rank.index, avg_by_rank["NumberOfPACs"], label="NumberOfPACs", color="Red")
plt.plot(avg_by_rank.index, avg_by_rank["SelectByHotkeys"], label="SelectByHotkeys", color="Gold")
plt.legend()
plt.title("High Correlation Variable Differences by Rank")
plt.xlabel("Rank")
plt.ylabel("Units")
plt.show()
```

One interesting takeaway from these graphs is that there are very big jumps from Grandmaster to Pro in APM and SelectByHotkeys.

### Modeling and Evaluation

Next, we need to train a multiclass classifier to predict the rank. I'm going to try a few different models: LogisticRegrsssion with Cross Validation, NearestNeighborClassification, and RandomForestClassification.

- LogisticRegression is generally a good model to always try.
- NearestNeighbor in this situation might be good because it may be the case that similar ranks are bunched up at certain values for a lot of variables.
- RandomForest are usually pretty good classifiers if the data has a lot of differences between classes.

Multi-class classification becomes more difficult as the number of classes increase, so let's see how these models perform.

```python
# Create Training dataset, Test dataset, and Validation Dataset
X_train, X_test, y_train, y_test, X_valid, y_valid = classifiers.get_data(starcraft, test_size=.3, random_state=42)
```

```python
# Find best LogReg regularization constant (C=__)
for i in [1, 5, 10, 15, 20]:
    curr_clf = classifiers.LogisticCV(X_train, y_train, i)
    classifiers.clf_results(curr_clf, X_test, y_test)
    print("\n")
```

```python
# Find best number of nearest neighbors
for i in [1, 3, 7, 15]:
    curr_clf = classifiers.KNeigbors(X_train, y_train, i)
    classifiers.clf_results(curr_clf, X_test, y_test)
    print("\n")
```

```python
# Find best decision tree max_depth parameters 
for i in [1, 5, 10, 13, 15]:
    curr_clf = classifiers.RandomForest(X_train, y_train, i)
    classifiers.clf_results(curr_clf, X_test, y_test)
    print("\n")
```

The RandomForestClassifier with a max_depth of 13 seems to be the best classifier as it's the only one that has above a 40% classification rate on the test data set. It also predicts within 1 rank 85% of the time which is not bad.

```python
# I wasn't fully satisfied with the 41% success rate and tried to see if I could use a Multi-layer Perceptron
# with OneVsRest to improve performance, but it was not better than the RandomForest.
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neural_network import MLPClassifier
clf = OneVsRestClassifier(MLPClassifier(hidden_layer_sizes = [100]*8, random_state=0)).fit(X_train, y_train)
classifiers.clf_results(clf, X_test, y_test)
```

Lastly, we test the Random Forest Classifier on the validation set to see it's performance. We end up with 39% correct predictions and 82.6% of predictions being within 1 rank.

```python
# As to be expected the validation numbers are slightly lower, but similar performance!
final_clf = classifiers.RandomForest(X_train, y_train, 13)
classifiers.clf_results(final_clf, X_valid, y_valid)
```

```python
# check most important features
final_clf.feature_importances_, final_clf.feature_names_in_
```

```python
# check for discrepancies in ranks that are commonly misclassfied
misclassifications = y_valid[final_clf.predict(X_valid) != y_valid].apply(lambda x: rank_dict[x]).value_counts()
misclassifications
```

```python
# Rank distribution again for reference 
starcraft["LeagueName"].value_counts()
```

```python
# Misclassfication proportions
(misclassifications / starcraft["LeagueName"].value_counts()).sort_values(ascending=False)
```

The most important features to the classifer are ActionLatency, APM, NumberofPACs, GapBetweenPACs, and SelectByHotkeys which basically emphasizes our earlier findings.

The model is misclassifying Diamond and Plat players the most, but that's also because they are more prevalent in the dataset. When looking at the ratios of misclassification, it shows we are misclassifying GrandMaster, Silver, and Gold players at high rates. One posititve though, is that generally we are predicting each class with an accuracy of 80% or higher.
